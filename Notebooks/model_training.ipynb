Imports the model on which the dataset have to be Trained 

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from catboost import CatBoostClassifier
```

Loading the dataset

spliting dataset into x and y

```python
X = df.drop(columns=['treatment','Age_Group'], axis=1) # features

y = df['treatment'] # target variable
```

```python
X.shape, y.shape
```

spliting the X,y in to X_train, X_test and y_train , y_test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
models = {
    'RandomForest': RandomForestClassifier(),
    'XGBoost': XGBClassifier(),
    'DecisionTree': DecisionTreeClassifier(),
    'LogisticRegression': LogisticRegression(),
    'SVC': SVC(),
    'GaussianNB': GaussianNB(),
    'CatBoostClassifier': CatBoostClassifier()
}
```

creating the function through which all models are trained and predict the output

```python
def train_and_evaluate(models, X_train, X_test, y_train, y_test):
    results = {}
    for model_name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        results[model_name] = accuracy
        print(f"Model: {model_name}")
        print(f"Accuracy: {accuracy:.4f}")
        print("Classification Report:")
        print(classification_report(y_test, y_pred))
        print("Confusion Matrix:")
        print(confusion_matrix(y_test, y_pred))
        print("-" * 50)
    return results
```

```python
train_and_evaluate(models, X_train, X_test, y_train, y_test)
```

# Now performing the hyperparameter tuning 
for the best result from the Logistic Regression model and CatBoostClassifier

to find max accuracy model for machine 

using the GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid_lr = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],  # Inverse of regularization strength
    'penalty': ['l1', 'l2'],            # Norm used in the penalization
    'solver': ['liblinear', 'saga']      # Algorithm to use in the optimization problem
}

print("Parameter grid for Logistic Regression:")
display(param_grid_lr)
```

```python
grid_search_lr = GridSearchCV(estimator=LogisticRegression(), param_grid=param_grid_lr,
                              scoring='accuracy', cv=5, n_jobs=-1, verbose=1)
grid_search_lr.fit(X_train, y_train)
best_lr = grid_search_lr.best_estimator_
y_pred_lr = best_lr.predict(X_test) 
accuracy_lr = accuracy_score(y_test, y_pred_lr)
print(f"Best Logistic Regression Model Accuracy: {accuracy_lr:.4f}")
print("Best Hyperparameters for Logistic Regression:")
print(grid_search_lr.best_params_)
```

By performing Hyperparameter tuning on the logistic Regression the 

Accuracy is now : 75 % 

the brest paremeters are : 'C': 0.1, 'penalty': 'l1', 'solver': 'saga'

Now performing the Hyperparameter tuning on the CatBoost classifier 

```python
param_grid_catboost = {
    'iterations': [100, 250, 500],  # Number of boosting iterations
    'learning_rate': [0.01, 0.05, 0.1],  # Learning rate
    'depth': [4, 6, 8],  # Depth of the trees
    'l2_leaf_reg': [1, 3, 5, 7],  # L2 regularization term
    'border_count': [32, 64, 128],  # The number of splits for numerical features
    'verbose': [False] # Suppress verbose output during training
}

print("Parameter grid for CatBoost Classifier:")
display(param_grid_catboost)
```

```python
grid_search_cb = GridSearchCV(estimator=CatBoostClassifier(), param_grid=param_grid_catboost,
                                scoring='accuracy', cv=5, n_jobs=-1, verbose=1)
grid_search_cb.fit(X_train, y_train)
best_cb = grid_search_cb.best_estimator_
y_pred_cb = best_cb.predict(X_test)
accuracy_cb = accuracy_score(y_test, y_pred_cb)
print(f"Best CatBoost Classifier Model Accuracy: {accuracy_cb:.4f}")
print("Best Hyperparameters for CatBoost Classifier:")
print(grid_search_cb.best_params_)
```

for HyperParemeter tuning in the CatBoost Classifier 

the best accuracy is  : 75 which is the higest 

the params are : {'border_count': 32, 'depth': 6, 'iterations': 100, 'l2_leaf_reg': 1, 'learning_rate': 0.01, 'verbose': False}

```python
param_grid_nb = {
    'var_smoothing': np.logspace(0, -9, num=100)
}

# Initialize GridSearchCV
grid_search_nb = GridSearchCV(estimator=GaussianNB(), 
                              param_grid=param_grid_nb,
                              scoring='accuracy', 
                              cv=5, 
                              n_jobs=-1, 
                              verbose=1)

# Fit the grid search to your data
grid_search_nb.fit(X_train, y_train)

# Get the best estimator
best_nb = grid_search_nb.best_estimator_

# Make predictions and evaluate
y_pred_nb = best_nb.predict(X_test) 
accuracy_nb = accuracy_score(y_test, y_pred_nb)

print(f"Best GaussianNB Model Accuracy: {accuracy_nb:.4f}")
print("Best Hyperparameters for GaussianNB:")
print(grid_search_nb.best_params_)
```
