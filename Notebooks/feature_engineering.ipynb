# New feature age group

```python
# Get the description of the age column to understand its distribution
age_description = df['Age'].describe()
print("Age Distribution Description:")
display(age_description)

# Based on the distribution, let's define some age bins.
# We can adjust these based on the describe output.
# For example, using quartiles or other logical groupings.
Q1 = df['Age'].quantile(0.25)
Q2 = df['Age'].quantile(0.50)
Q3 = df['Age'].quantile(0.75)

# Let's define bins based on quartiles and add some boundaries
age_bins = [df['Age'].min(), Q1, Q2, Q3, df['Age'].max()]
age_labels = ['Group 1', 'Group 2', 'Group 3', 'Group 4'] # Generic labels, can be customized

# Create a new column 'Age_Group' by categorizing 'Age' using the defined bins
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, include_lowest=True)

# Display the first few rows with the new 'Age_Group' column and the count of each group
display(df.head())
print("\nDistribution of Age Groups:")
display(df['Age_Group'].value_counts())
```
