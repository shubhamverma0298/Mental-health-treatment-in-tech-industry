# checking outliers in age column

as a result their are some outliers in the data set 


```python
Q1 = dff['Age'].quantile(0.25)
Q3 = dff['Age'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_no_outliers = dff[(dff['Age'] >= lower_bound) & (dff['Age'] <= upper_bound)]

print(f"Original number of rows: {len(dff)}")
print(f"Number of rows after removing outliers: {len(df_no_outliers)}")

# Display the first few rows of the new dataframe without outliers
display(df_no_outliers.head())
```

# Perform exploratory data analysis (EDA) on the dataset to understand its characteristics, including checking for missing values, visualizing distributions, and exploring relationships between features.

```python
# Histogram for 'Age'
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'].dropna(), kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Count plot for 'Gender'
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df)
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

0 for male 

1 for female 

2 for others 

```python
# 4. Count plots for other relevant categorical columns
categorical_cols = ['self_employed', 'family_history', 'treatment', 'work_interfere',
                    'remote_work', 'tech_company', 'benefits', 'care_options',
                    'wellness_program', 'seek_help', 'anonymity', 'leave',
                    'mental_health_consequence', 'coworkers', 'supervisor',
                    'mental_health_interview', 'mental_vs_physical', 'obs_consequence',
                    'phys_health_consequence','phys_health_interview']

for col in categorical_cols:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
#  Generate a correlation matrix for numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numerical_cols].corr()
```

```python
# Display the correlation matrix as a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

for Treatment 'Yes':1,'No':0

```python
# Create grouped bar plots for categorical features vs treatment
categorical_cols_for_eda = ['Gender', 'family_history', 'work_interfere',
                            'remote_work', 'tech_company', 'benefits', 'care_options',
                            'wellness_program', 'seek_help', 'anonymity', 'leave',
                            'mental_health_consequence', 'coworkers', 'supervisor',
                            'mental_health_interview', 'mental_vs_physical', 'obs_consequence'
                            'phys_health_consequence','phys_health_interview']

for col in categorical_cols_for_eda:
    if col in df.columns:
        sns.catplot(x=col, hue='treatment', kind='count', data=df, height=6, aspect=1.5)
        plt.title(f'Relationship between {col} and Treatment')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.show()
```
